```python
from glob import glob
import pandas as pd
```

# Demographic information

### Training Set A

```python
training_setA = []

files = glob('training_setA/*.csv')
```

```python
for file in files:
    patient = file.split('\\')[1].split('.')[0]
    train_set = 'A'
    
    ## Reading the patient data
    data = pd.read_csv(file,sep = '|')
    
    ## Getting demographic and unit info
    age = data.loc[0,'Age']

    if data.loc[0,'Gender']==0:
        gender = 'Female'
    else:
        gender = 'Male'

    ## NAs also regarded as 0
    MICU_unit = int(any(data['Unit1']==1))
    SICU_unit = int(any(data['Unit2']==1))

    ## Assuming this field doesn't change for a patient
    HospAdmTime = data.loc[0,'HospAdmTime']

    ICULOS = data.iloc[-1,39]        #We can also use length of data

    Sepsis = int(any(data['SepsisLabel']==1))
    
    ## Appending the info to list
    training_setA.append([train_set, patient,age, gender,
                           MICU_unit, SICU_unit, HospAdmTime, ICULOS, Sepsis])
```

```python
training_setA = pd.DataFrame(training_setA,columns = ["Train_set", "Patient","Age", "Gender",
                           "MICU_unit", "SICU_unit", "HospAdmTime", "ICULOS", "Sepsis"]).copy()
training_setA
```

### Training Set B

```python
training_setB = []

files = glob('training_setB/*.csv')
```

```python
for file in files:
    patient = file.split('\\')[1].split('.')[0]
    train_set = 'B'
    
    ## Reading the patient data
    data = pd.read_csv(file,sep = '|')
    
    ## Getting demographic and unit info
    age = data.loc[0,'Age']

    if data.loc[0,'Gender']==0:
        gender = 'Female'
    else:
        gender = 'Male'

    ## NAs also regarded as 0
    MICU_unit = int(any(data['Unit1']==1))
    SICU_unit = int(any(data['Unit2']==1))

    ## Assuming this field doesn't change for a patient
    HospAdmTime = data.loc[0,'HospAdmTime']

    ICULOS = data.iloc[-1,39]        #We can also use length of data

    Sepsis = int(any(data['SepsisLabel']==1))
    
    ## Appending the info to list
    training_setB.append([train_set, patient,age, gender,
                           MICU_unit, SICU_unit, HospAdmTime, ICULOS, Sepsis])
```

```python
training_setB = pd.DataFrame(training_setB,columns = ["Train_set", "Patient","Age", "Gender",
                           "MICU_unit", "SICU_unit", "HospAdmTime", "ICULOS", "Sepsis"]).copy()
training_setB
```

### combining both

```python
training_setall = pd.concat([training_setA, training_setB],ignore_index=True)
training_setall
```

```python
training_setall.to_csv("trains.csv",index=False)
```
